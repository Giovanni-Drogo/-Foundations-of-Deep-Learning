```python
##  ==============================
from PIL import Image
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# 
image0 = Image.open("/mnt/workspace///Monkey D Luffy.webp")

# 
num = 10

my_transform1 = transforms.RandomVerticalFlip(p=0.5)
my_transform2 = transforms.RandomHorizontalFlip(p=0.5)
my_transform3 = transforms.RandomResizedCrop(
    size=200,         # 
    scale=(0.1, 1.0), # 
    ratio=(0.5, 2),   # 
    interpolation=2,  # 2
)  
my_transform4 = transforms.ColorJitter(
    brightness=0.5, # 
    contrast=0.5,   # 
    saturation=0.5, # 
    hue=0.5)        # 

combined_transform = transforms.Compose(
    [   my_transform1,
        my_transform2,
        my_transform3,
        my_transform4  ]  )

# 
image = []

# num
for i in range(num):
    image.append(combined_transform(image0))

# ----------
# 
plt.figure(figsize=(20, 10)) # (, ), 

# 
for i in range(num):
    plt.subplot(1, num, i + 1)
    plt.imshow(image[i])
    plt.axis("off")

# 
plt.show()
```
