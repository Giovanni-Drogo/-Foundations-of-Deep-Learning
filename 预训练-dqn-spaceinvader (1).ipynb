```python
from stable_baselines3 import DQN
import gymnasium as gym

# 创建与训练时相同的环境（参数必须一致）
env = gym.make("ALE/SpaceInvaders-v5", max_episode_steps=1000, render_mode="human")

# 加载模型（文件在当前目录下的 `ale_space_invaders` 文件夹）
model = DQN.load("ale_space_invaders", env=env)

# 测试模型性能（运行一个回合）
obs, _ = env.reset()
for _ in range(1000):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, _, _ = env.step(action)
    if done:
        break
env.close()
```

```python
import gymnasium as gym
import ale_py
from stable_baselines3 import DQN

gym.register_envs(ale_py)  # 注册 ALE 环境
env = gym.make("ALE/SpaceInvaders-v5", render_mode="human")
model = DQN.load("ale_space_invaders")

obs, info = env.reset()
total_reward = 0
while True:
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = env.step(int(action))
    total_reward += reward
    if terminated or truncated:
        obs, info = env.reset()
        print(f"Episode done, reward: {total_reward}")
        print(f"Episode done, truncated: {truncated}, terminated: {terminated}")
        break  # exit(0)
```

```python
import gymnasium as gym
from stable_baselines3 import DQN

# 创建与训练时一致的环境（参数必须完全匹配）
env = gym.make(
    "ALE/SpaceInvaders-v5", 
    max_episode_steps=1000,  # 确保与预训练时一致
    render_mode=None          # 训练时不渲染
)

# 加载预训练模型（假设文件在 ale_space_invaders 文件夹下）
model = DQN.load("ale_space_invaders", env=env)

# 继续训练（追加 1e5 步）
model.learn(
    total_timesteps=int(1e5),
    log_interval=50,          # 每50步输出一次日志
    reset_num_timesteps=False # 不重置步数计数器（保留原训练步数）
)

# 保存更新后的模型
model.save("dqn_spaceinvaders_continued")
```

```python
import gymnasium as gym
from stable_baselines3 import DQN

# 创建测试环境（允许渲染）
test_env = gym.make(
    "ALE/SpaceInvaders-v5", 
    max_episode_steps=1000,
    render_mode="human"  # 测试时显示画面
)

# 加载预训练后的模型
model = DQN.load("ale_space_invaders", env=test_env)

# 运行10个回合并统计平均奖励
total_rewards = []
for _ in range(10):
    obs, _ = test_env.reset()
    episode_reward = 0
    done = False
    while not done:
        action, _ = model.predict(obs, deterministic=True)  # 确定性策略
        obs, reward, done, _, _ = test_env.step(action)
        episode_reward += reward
    total_rewards.append(episode_reward)
    test_env.close()

print(f"平均奖励（10回合）: {sum(total_rewards)/len(total_rewards):.1f}")
```

```python
import gymnasium as gym
from stable_baselines3 import DQN

# 创建测试环境（允许渲染）
test_env = gym.make(
    "ALE/SpaceInvaders-v5", 
    max_episode_steps=1000,
    render_mode="human"  # 测试时显示画面
)

# 加载继续训练后的模型
model = DQN.load("dqn_spaceinvaders_continued", env=test_env)

# 运行10个回合并统计平均奖励
total_rewards = []
for _ in range(10):
    obs, _ = test_env.reset()
    episode_reward = 0
    done = False
    while not done:
        action, _ = model.predict(obs, deterministic=True)  # 确定性策略
        obs, reward, done, _, _ = test_env.step(action)
        episode_reward += reward
    total_rewards.append(episode_reward)
    test_env.close()

print(f"平均奖励（10回合）: {sum(total_rewards)/len(total_rewards):.1f}")
```

```python
def evaluate_model(model_path, env_name, num_episodes=10):
    env = gym.make(env_name, max_episode_steps=1000)
    model = DQN.load(model_path, env=env)
    total_rewards = []
    for _ in range(num_episodes):
        obs, _ = env.reset()
        done = False
        episode_reward = 0
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, done, _, _ = env.step(action)
            episode_reward += reward
        total_rewards.append(episode_reward)
    env.close()
    return sum(total_rewards) / num_episodes

# 对比预训练和继续训练后的模型
original_score = evaluate_model("ale_space_invaders", "ALE/SpaceInvaders-v5")
continued_score = evaluate_model("dqn_spaceinvaders_continued", "ALE/SpaceInvaders-v5")
print(f"预训练模型平均奖励: {original_score:.1f}")
print(f"继续训练后平均奖励: {continued_score:.1f}")
```
