```python
import os

# 设置从https://hf-mirror.com下载模型，否则会从huggingface.co下载
os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"

from transformers import pipeline

# 下载模型
translator = pipeline("translation_en_to_zh",   # 英文翻译为中文
                      model="Helsinki-NLP/opus-mt-en-zh", # 下载模型到本地
                      device=0)        # 使用第一个GPU, 去掉该参数则使用CPU

```

```python

# 英文
text = 'I offer you the loyalty of a man who has never been loyal.'
# 翻译
translation = translator(text)

# 打印翻译结果
print(translation)
print(translation[0].get('translation_text'))
```
