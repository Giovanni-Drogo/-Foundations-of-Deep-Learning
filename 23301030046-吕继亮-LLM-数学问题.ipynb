```python
import openai

client = openai.OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")


def solve(problem):
    SYSTEM_MESSAGE = f"""
    你是一个数字密码破解专家。
    目标是用给定四个数字和基本运算，计算出最接近100的结果（可超过）。
    每个数字必须使用一次且只能使用一次。请制定一个解题计划，逐步解释你的思路，
    请输出表达式和最终值。
    """
    

    PROBLEM = f"""给定以下数字，请输出一个表达式和最终值:\n{problem}"""

    chat_response = client.chat.completions.create(
        model="smallthinker:latest",
        messages=[
            {"role": "system", "content": f"""{SYSTEM_MESSAGE}"""},
            {"role": "user", "content": f"""{PROBLEM}"""},
        ],
        temperature=0.7,
        top_p=1.0,
        max_tokens=4096,
        extra_body={"repetition_penalty": 1.15},
    )
    # print("Chat response:", chat_response)
    return chat_response.choices[0].message.content


print(solve("[4, 3, 7, 25]"))


```

```python
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = Ollama(
    model="smallthinker:latest",
    base_url="http://localhost:11434",
    temperature=0.7,
    top_p=0.9
)
SYSTEM_MESSAGE = """
你是一个数字密码破解专家。
目标是用给定四个数字和基本运算，计算出最接近100的结果（可超过）。
每个数字必须使用一次且只能使用一次。
请你分步思考，先分析数字组合的可能性，再尝试不同的运算顺序，最后给出最优表达式和最终值。
请严格按照以下格式输出：
解题思路：
1. ...
2. ...
最终表达式：...
最终值：...3
"""
prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_MESSAGE),
    ("human", "给定数字：[3, 4, 7, 25]，请解答。")
])

chain = prompt | llm | StrOutputParser()
result = chain.invoke({})
print(result)
```

```python
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate, FewShotChatMessagePromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = Ollama(model="smallthinker:latest", base_url="http://localhost:11434")

# 1. 定义示例
examples = [
    {
        "problem": "给定数字 [7, 8, 9, 10]",
        "solution": "10 * (9 + (8 - 7)) = 100"
    },
    {
        "problem": "给定数字 [2, 5, 6, 12]",
        "solution": "(12 + 6) * (5 + 2) = 126"
    },
    {
        "problem": "给定数字 [1, 2, 3, 16]",
        "solution": "(16)*(3+2+1)=96"
    }
    {        "problem": "给定数字 [3, 4, 7, 20]",
        "solution": "(20 + 4) * (7 - 3) = 96"
    },
    {
        "problem": "给定数字 [5, 6, 8, 15]",
        "solution": "(5 +8 ) * (15-6)  = 117"
        
    }
]

# 2. 定义 few-shot 提示模板
example_prompt = ChatPromptTemplate.from_messages([
    ("human", "{problem}"),
    ("ai", "{solution}")
])

few_shot_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples
)

# 3. 组合完整提示
full_prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个数字密码破解专家。目标是用给定四个数字和基本运算，计算出最接近100的结果。每个数字必须使用一次且只能使用一次。请直接输出表达式和最终值。"),
    few_shot_prompt,
    ("human", "{problem}")
])

# 4. 构建链
chain = full_prompt | llm | StrOutputParser()

# 5. 调用
result = chain.invoke({"problem": "给定数字 [3, 4, 7, 25]"})
print(result)
```

```python
from langchain_community.llms import Ollama
from langchain.chains import LLMMathChain
from langchain.tools import Tool
from langchain.agents import initialize_agent, AgentType

# 初始化模型
llm = Ollama(model="smallthinker:latest", base_url="http://localhost:11434",
             temperature=0.7, top_p=0.9)

# 初始化数学链
math_chain = LLMMathChain.from_llm(llm=llm, verbose=True)

# 定义计算器工具
calculator_tool = Tool(
    name="Calculator",
    func=math_chain.run,
    description="用于数学计算的工具。输入应为完整的数学问题。"
)

# 初始化 Agent
tools = [calculator_tool]
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# 提问
query = "给定数字 [3, 4, 7, 25]，计算出最接近100的表达式。"
response = agent.invoke({"input": query}).get("output", "No output found.")
print(response)
```

```python
import itertools
import operator

def closest_to_100(nums):
    ops = [('+', operator.add), ('-', operator.sub), ('*', operator.mul), ('/', operator.truediv)]
    best = None
    best_expr = ""
    for nums_perm in itertools.permutations(nums):
        for ops_combo in itertools.product(ops, repeat=3):
            exprs = [
                f"(({nums_perm[0]}{ops_combo[0][0]}{nums_perm[1]}){ops_combo[1][0]}{nums_perm[2]}){ops_combo[2][0]}{nums_perm[3]}",
                f"({nums_perm[0]}{ops_combo[0][0]}({nums_perm[1]}{ops_combo[1][0]}{nums_perm[2]})){ops_combo[2][0]}{nums_perm[3]}",
                f"{nums_perm[0]}{ops_combo[0][0]}(({nums_perm[1]}{ops_combo[1][0]}{nums_perm[2]}){ops_combo[2][0]}{nums_perm[3]})",
                f"{nums_perm[0]}{ops_combo[0][0]}({nums_perm[1]}{ops_combo[1][0]}({nums_perm[2]}{ops_combo[2][0]}{nums_perm[3]}))"
            ]
            for expr in exprs:
                try:
                    val = eval(expr)
                    if best is None or abs(val-100) < abs(best-100):
                        best = val
                        best_expr = expr
                except ZeroDivisionError:
                    continue
    return best_expr, best

expr, val = closest_to_100([3, 4, 7, 25])
print(f"{expr} = {val}")
```
